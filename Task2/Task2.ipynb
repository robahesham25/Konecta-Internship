## Multiple pages and multiple products

```python
products_dict = {'title': [],'brand': [], 'price': [], 'rating': [] , 'review_count': [],'product_link':[], 'image_url': []}
```

```python
def get_title(soup):
    try: 
        title = product_soup.find('span' , attrs = {'class': 'a-size-large product-title-word-break'}).text.strip()
    except AttributeError:
        title = ''
    return title

def get_price(soup):
     try: 
        price = product_soup.find('span', attrs = {'class': 'a-offscreen'}).text
     except AttributeError:
        price = ''
     return price

def get_brand(soup):
     try: 
        brand = product_soup.find('span', attrs = {'class': 'a-size-base po-break-word'}).text
     except AttributeError:
        brand = ''
     return brand

def get_rating(soup):
    try:
        rating = product_soup.find('span', attrs = {'class': 'a-icon-alt'}).text.split()[0]
    except AttributeError:
        rating = ''
    return rating

def get_reviews(soup):
     try:
        reviews = product_soup.find('span', attrs = {'id' :"acrCustomerReviewText"}).text.split()[0]
     except AttributeError:
        reviews = ''
     return reviews

def get_image(soup):
     try:
        image = product_soup.find('div', attrs = {'class': 'imgTagWrapper'}).img.get('src')
     except AttributeError:
        image = ''
     return image
```

```python
for page in range(1,2):
    url = 'https://www.amazon.com/s?k=laptops&crid=1MMTRW2O10Y0J&qid=1754655257&sprefix=laptops%2Caps%2C264&xpid=V4OMaF1xdb4Ig&ref=sr_pg_'+ str(page)
    response = requests.get(url , headers = headers)
    soup = BeautifulSoup(response.content, 'html.parser')
    products = soup.find_all('a', attrs = {'class': 'a-link-normal s-line-clamp-2 s-link-style a-text-normal'})
    products_list = []
    for product in products:
        products_list.append(root + product.get('href'))
    for product in products_list:
        webpage = requests.get(product, headers = headers)
        product_soup = BeautifulSoup(webpage.content, 'html.parser')
        products_dict['title'].append(get_title(product_soup))
        products_dict['brand'].append(get_brand(product_soup))
        products_dict['price'].append(get_price(product_soup))
        products_dict['rating'].append(get_rating(product_soup))
        products_dict['review_count'].append(get_reviews(product_soup))
        products_dict['product_link'].append(product)
        products_dict['image_url'].append(get_image(product_soup))
```

```python
amazon_df = pd.DataFrame.from_dict(products_dict)
amazon_df
```

```python
amazon_df = amazon_df.replace(r'^\s*$', np.nan, regex=True)
```
